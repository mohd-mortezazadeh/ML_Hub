```python
import matplotlib.pyplot as plt
x = ["red", "green", "yellow", "blue"]
y = [5,6,7,8]
plt.pie(y, labels=x, autopct="%1.1f%%",colors=["red", "green", "yellow", "blue"])
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["red", "green", "yellow", "blue"]
y = [5,6,7,8]
plt.bar(x, y, color="green")
plt.show()
```
