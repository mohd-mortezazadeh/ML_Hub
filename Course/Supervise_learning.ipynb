```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LogisticRegression


x = np.array([1,2,3,4]).reshape(-1, 1)
y = np.array([10,20,30,40])
model = LinearRegression()
model.fit(x, y)
m = model.coef_[0]
b = model.intercept_

plt.plot(x, y, marker="*", color="red")
plt.plot(x, model.predict(x))
plt.show()
result = model.predict([[5]])
print(result)
print(m, b)
```

# # Sample data

```python


# Sample data
data = {
    'age': [25, 30, 35, 40, 45],
    'income': [50000, 60000, 70000, 80000, 90000]
}
df = pd.DataFrame(data)

# Features and target variable
x = df[['age']]  # Reshape to 2D array
y = df['income']

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Create and train the model
model = LinearRegression()
model.fit(x_train, y_train)

# Make predictions
y_pred = model.predict(x_test)

# Optionally, evaluate the model
mse = mean_squared_error(y_test, y_pred) #خطای میانگین مربعات یک معیار ارزیابی برای مدل‌های رگرسیون است
r2 = r2_score(y_test, y_pred) 
#R² Score یا Coefficient of Determination
#  یک معیار ارزیابی برای مدل‌های رگرسیون است که نشان می‌دهد چه درصدی از واریانس متغیر وابسته (هدف) توسط متغیرهای مستقل (ویژگی‌ها) توضیح داده می‌شود.
# Print metrics
print("Mean Squared Error:", mse)
print("R^2 Score:", r2)

# Plotting
plt.scatter(x, y, color='green', label="real data")

# Create a range of values for the regression line
x_range = np.linspace(x['age'].min(), x['age'].max()).reshape(-1, 1)
y_range = model.predict(x_range)

plt.plot(x_range, y_range, color='red', label="linear regression")
plt.xlabel('Age')
plt.ylabel('Income')
plt.title("Linear Regression")
plt.legend()
plt.show()
```

# Polynomial Regression 

```python
#  رگرسیون چندجمله‌ای 
from sklearn.preprocessing import PolynomialFeatures

# 1. Generate nonlinear data
np.random.seed(0)  # Set seed for reproducibility
X = np.linspace(-3, 3, 100).reshape(-1, 1)  # Create 100 data points in the range [-3, 3]
y = X**2 + np.random.normal(0, 0.5, X.shape)  # Nonlinear function (x^2) with added noise

# 2. Transform the data into new polynomial features
degree = 2  # Set the degree of the polynomial
poly_features = PolynomialFeatures(degree=degree)  # Create a PolynomialFeatures object
X_poly = poly_features.fit_transform(X)  # Transform the original features into polynomial features

# 3. Apply linear regression on the transformed data
model = LinearRegression()  # Create a LinearRegression model
model.fit(X_poly, y)  # Fit the model to the polynomial features and target variable

# 4. Predict based on the model
y_pred = model.predict(X_poly)  # Make predictions using the fitted model

# 5. Display the results
plt.scatter(X, y, color='blue', label='Actual Data')  # Plot the actual data points
plt.plot(X, y_pred, color='red', label='Polynomial Regression')  # Plot the predicted values
plt.title('Linear Regression on Nonlinear Data')  # Set the title of the plot
plt.xlabel('X')  # Label for the x-axis
plt.ylabel('y')  # Label for the y-axis
plt.legend()  # Show the legend
plt.grid()  # Show grid lines on the plot
plt.show()  # Display the plot
```

(پردازش زبان طبیعی و تحلیل متن)


**شکل منحنی:**

 توزیع گوسی به صورت یک منحنی زنگی شکل است که مرکز آن میانگین (μ) و عرض آن انحراف معیار (σ) را نشان می‌دهد.

**تقارن:**

 این توزیع نسبت به میانگین خود تقارن دارد، به این معنی که مقادیر بالاتر و پایین‌تر از میانگین به یک اندازه احتمال دارند.
 
**قانون زنگوله‌ای:**

 تقریباً 68% از داده‌ها در یک انحراف معیار از میانگین، 95% در دو انحراف معیار و 99.7% در سه انحراف معیار قرار دارند (قانون 68-95-99.7).

- Gaussian Naive Bayes: برای داده‌های پیوسته با توزیع نرمال.
- Multinomial Naive Bayes: برای داده‌های گسسته و متنی.

مزایا:
- ساده و سریع: الگوریتمی ساده و سریع است که به راحتی پیاده‌سازی می‌شود.

معایب:

# PCA

![My Image](../images/principal-component-analysis-pca-featured.png)



ابعاد یک مجموعه داده را با حفظ بیشترین واریانس ممکن کاهش می‌دهد

کاهش پیچیدگی و بهبود عملکرد مدل‌ها کمک می‌کند.

مزایا:

**کاهش نویز:**

- با حذف ابعاد کم‌اهمیت می‌تواند به کاهش نویز در داده‌ها کمک کند

**بهبود کارایی:**

- کاهش ابعاد می‌تواند زمان و منابع محاسباتی را کاهش دهد.

محور افقی (اولین مؤلفه اصلی) باید بیشترین واریانس داده‌ها را حفظ کند

محور عمودی (دومین مؤلفه اصلی) باید کمترین واریانس از دست رفته را داشته باشد یا 

 کمترین اطلاعات مهم از داده‌ها از دست می‌رود




```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
pca.fit(X_train)
X_new_train = pca.transform(X_train)
X_new_test = pca.transform(X_test)
X_train.shape, X_new_train.shape
```

```python
model = MLPClassifier(hidden_layer_sizes=64, max_iter=200)
model.fit(X_new_train, y_train)
y_new_pred_train = model.predict(X_new_train)
y_new_pred_test = model.predict(X_new_test)

acc_train = accuracy_score(y_true=y_train, y_pred=y_new_pred_train)
acc_test = accuracy_score(y_true=y_test, y_pred=y_new_pred_test)
acc_train, acc_test
```
